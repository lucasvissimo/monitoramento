```python
import requests

API_KEY = "ELFkM8LTsLpTmbvzVyVqCslVDfBdNAcP"
H = {"X-EVINO-KESTRA-API-KEY": API_KEY}
BASE = "https://api.evino.com.br/kestra/api/v1/main"

# pega flows
flows = requests.get(f"{BASE}/flows/rpa.varejofacil", headers=H).json()

for f in flows:
    full_id = f"{f['namespace']}.{f['id']}"
    url = f"{BASE}/executions?flowId={full_id}&size=1&sort=startedAt:DESC"
    r = requests.get(url, headers=H)
    if r.ok:
        items = r.json().get("results", r.json())  # dependendo do formato
        latest = items[0] if items else None
        print(full_id, "→", latest and latest.get("state", {}).get("current"), latest and latest.get("startDate"))
    else:
        print(full_id, "→ erro", r.status_code, r.text[:200])
```

```python
import requests
from urllib.parse import quote

API_KEY = "ELFkM8LTsLpTmbvzVyVqCslVDfBdNAcP"
HEADERS = {"X-EVINO-KESTRA-API-KEY": API_KEY}
BASE_URL = "https://api.evino.com.br/kestra/api/v1/main"
NAMESPACE = "rpa.varejofacil"

flow_ids = [
    "rpa_vf_clientes_ev",
    "rpa_vf_clientes_franquias_ev",
    "rpa_vf_clientes_franquias_gc",
    "rpa_vf_clientes_gc",
    "rpa_vf_cupons_ev",
    "rpa_vf_cupons_ev_teste",
    "rpa_vf_cupons_franquias_ev",
    "rpa_vf_cupons_franquias_ev_reprocessamento",
    "rpa_vf_cupons_franquias_gc",
    "rpa_vf_cupons_franquias_gc_reprocessamento",
    "rpa_vf_cupons_lojas_gc",
    "rpa_vf_cupons_lojas_gc_teste",
    "rpa_vf_relatorio_ABCVendas_ev",
    "rpa_vf_relatorio_desconto_gc",
]

for fid in flow_ids:
    
    full_id = f"{NAMESPACE}.{fid}"
    url = f"{BASE_URL}/executions?namespace={NAMESPACE}&flowId={quote(full_id)}&size=1&sort=startedAt:DESC"
    resp = requests.get(url, headers=HEADERS)

    if resp.ok:
        data = resp.json()
        
        executions = data.get("results", data if isinstance(data, list) else [])
        if executions:
            last = executions[0]
            status = last.get("state", {}).get("current")
            start_date = last.get("startDate")
            print(f"{fid:<40} → {status} @ {start_date}")
        else:
            print(f"{fid:<40} → nenhum histórico de execuções")
    else:
        print(f"{fid:<40} → erro {resp.status_code}: {resp.text[:150]}")
```
